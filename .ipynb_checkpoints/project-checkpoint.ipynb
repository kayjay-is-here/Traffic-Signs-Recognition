```python
from keras.preprocessing.image import ImageDataGenerator

train_dir = "data/Train"
test_dir = "data/Test"

train_images = ImageDataGenerator(rescale=1./255).flow_from_directory(
    train_dir,
    class_mode="categorical",
    target_size=(30,30),
    batch_size=32)
```

```python
from keras import utils
import numpy as np
import os
from tqdm.auto import tqdm

def read_images(dataframe):

    color_images = []
    grayscale_images = []
    dataframe.dropna(inplace=True)
    pbar = tqdm(total=dataframe.shape[0])

    for index, row in dataframe.iterrows():
        colored_image = utils.load_img(os.path.join("data", row["Path"]), target_size=(30,30), color_mode="rgb")
        grayscale_image = utils.load_img(os.path.join("data", row["Path"]), target_size=(30,30), color_mode="grayscale")

        color_images.append(np.asarray(utils.img_to_array(colored_image), dtype="float32"))
        grayscale_images.append(np.asarray(utils.img_to_array(grayscale_image), dtype="float32"))

        pbar.update(1)

    pbar.close()
    return grayscale_images, color_images
```

```python
train_data = {}
test_data = {}
train_data["grayscale"], train_data["color"] = read_images(train_data_label)
test_data["grayscale"], test_data["color"] = read_images(test_data_label)
```

```python
import keras.backend as K
def get_f1(y_true, y_pred): #taken from old keras source code
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val
```

```python
import tensorflow as tf
from keras import layers
from keras import metrics

FILTER_COUNT = 64
FILTER_SIZE = (3,3) # 3x3 filters

grayscale_model = tf.keras.models.Sequential()
grayscale_model.add(layers.Conv2D(FILTER_COUNT, FILTER_SIZE, activation='relu', input_shape=(30,30,1)))

grayscale_model.add(layers.MaxPooling2D(pool_size=(2,2)))
#color_model.add(layers.GlobalAveragePooling2D())
grayscale_model.add(layers.Dropout(0.25))

grayscale_model.add(layers.Conv2D(FILTER_COUNT, FILTER_SIZE, activation='relu'))
grayscale_model.add(layers.Conv2D(FILTER_COUNT, FILTER_SIZE, activation='relu'))

grayscale_model.add(layers.Dropout(0.5)) # percentage of random neurons that are dropped each epoch
grayscale_model.add(layers.Conv2D(FILTER_COUNT, FILTER_SIZE, activation='relu'))

# color_model.add(layers.Dense(256, activation='relu')) # Remove this line if it's not helping
#color_model.add(layers.Dropout(0.5)) # Remove this line if it's not helping
grayscale_model.add(layers.Flatten())
grayscale_model.add(layers.Dense(43, activation='softmax')) # numbers 0 - 42 represent the
#color_model.add(layers.Dense(43, activation='softmax'))
# categorical labels

# Use ADAM optimizer
LEARNING_RATE = 0.001 # Try 0.001 next

adam = tf.keras.optimizers.Adam(
    learning_rate=LEARNING_RATE
)

grayscale_model.compile(loss="sparse_categorical_crossentropy", optimizer=adam, metrics=metrics.SparseCategoricalAccuracy())
```

```python
import tensorflow as tf
from keras import layers
from keras import metrics

FILTER_COUNT = 64
FILTER_SIZE = (3,3) # 3x3 filters

color_model = tf.keras.models.Sequential()
color_model.add(layers.Conv2D(FILTER_COUNT, FILTER_SIZE, activation='relu', input_shape=(30,30,3)))

color_model.add(layers.MaxPooling2D(pool_size=(2,2)))
#color_model.add(layers.GlobalAveragePooling2D())
color_model.add(layers.Dropout(0.25))

color_model.add(layers.Conv2D(FILTER_COUNT, FILTER_SIZE, activation='relu'))
color_model.add(layers.Conv2D(FILTER_COUNT, FILTER_SIZE, activation='relu'))

color_model.add(layers.Dropout(0.5)) # percentage of random neurons that are dropped each epoch
color_model.add(layers.Conv2D(FILTER_COUNT, FILTER_SIZE, activation='relu'))

# color_model.add(layers.Dense(256, activation='relu')) # Remove this line if it's not helping
#color_model.add(layers.Dropout(0.5)) # Remove this line if it's not helping
color_model.add(layers.Flatten())
color_model.add(layers.Dense(43, activation='softmax')) # numbers 0 - 42 represent the
#color_model.add(layers.Dense(43, activation='softmax'))
# categorical labels

# Use ADAM optimizer
LEARNING_RATE = 0.001 # Try 0.001 next

adam = tf.keras.optimizers.Adam(
    learning_rate=LEARNING_RATE
)

color_model.compile(loss="sparse_categorical_crossentropy", optimizer=adam, metrics=[metrics.SparseCategoricalAccuracy(),get_f1])
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_sparse_categorical_accuracy', patience=2) # change monitor to val_accuracy if this doesn't work
```

```python
X_train = np.asarray(train_data["color"], dtype=np.float32)

X_train = X_train.reshape(-1, 30, 30, 3)
y_train = train_data_label["ClassId"]

X_test = np.asarray(test_data["color"], dtype=np.float32)
X_test = X_test.reshape(-1, 30, 30, 3)
y_test = test_data_label["ClassId"]

color_data = {
    "X_train": X_train,
    "y_train": y_train,
    "X_test": X_test,
    "y_test": y_test
}

X_train = np.asarray(train_data["grayscale"], dtype=np.float32)

X_train = X_train.reshape(-1, 30, 30, 1)
y_train = train_data_label["ClassId"]

X_test = np.asarray(test_data["grayscale"], dtype=np.float32)
X_test = X_test.reshape(-1, 30, 30, 1)
y_test = test_data_label["ClassId"]

grayscale_data = {
    "X_train": X_train,
    "y_train": y_train,
    "X_test": X_test,
    "y_test": y_test
}
```

```python
%%time

np.shape(color_data["X_train"])
color_history = color_model.fit(
    color_data["X_train"],
    color_data["y_train"],
    epochs=5,
    validation_data=(color_data["X_test"], color_data["y_test"]),
    callbacks=[early_stop],
    verbose=True,
    batch_size=64
)
```

```python
%%time

np.shape(grayscale_data["X_train"])
grayscale_history = grayscale_model.fit(
    grayscale_data["X_train"],
    grayscale_data["y_train"],
    epochs=5,
    validation_data=(grayscale_data["X_test"], grayscale_data["y_test"]),
    callbacks=[early_stop],
    verbose=True,
    batch_size=64
)
```

```python
grayscale_acc = grayscale_history.history['sparse_categorical_accuracy']
grayscale_val_acc = grayscale_history.history['val_sparse_categorical_accuracy']
grayscale_loss = grayscale_history.history['loss']
grayscale_val_loss = grayscale_history.history['val_loss']

epochs_range = range(5)

plt.figure(figsize=(8, 8))
plt.subplot(2, 2, 2)
plt.plot(epochs_range, grayscale_acc, label='Training Accuracy')
plt.plot(epochs_range, grayscale_val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(2, 2, 2)
plt.plot(epochs_range, grayscale_loss, label='Training Loss')
plt.plot(epochs_range, grayscale_val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss For Color')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve
from sklearn.preprocessing import label_binarize

# Binarize the label
y_true = np.array(color_data['y_test'])
n_classes = 43
y_true = np.eye(n_classes)[y_true]

# Calculate the prediction probabilities for each class
y_pred = np.array(color_history.model.predict(color_data['X_test']))

```

```python
# Compute the average precision score for each class
precision = dict()
recall = dict()
average_precision = dict()
for i in range(n_classes):
    precision[i], recall[i], _ = precision_recall_curve(y_true[:, i],
                                                        y_pred[:, i])
    average_precision[i] = np.average(precision[i])

# Plot the PR curve for each class
colors = plt.cm.rainbow(np.linspace(0, 1, 42))
for i, color in zip(range(42), colors):
    plt.plot(recall[i], precision[i], color=color,
             label='{0} (AP = {1:0.2f})'
                   ''.format(class_names[i], average_precision[i]))

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall curve')
plt.legend(loc="center left", bbox_to_anchor=(1.05, 0.5), fontsize=8)
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true[:, i], y_pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot the ROC curve for each class
fig2, ax2 = plt.subplots()
for i, color in zip(range(n_classes), colors):
    ax2.plot(fpr[i], tpr[i], color=color,
             label='{0} (AUC = {1:0.2f})'
                   ''.format(class_names[i], roc_auc[i]))

ax2.plot([0, 1], [0, 1], 'k--')
ax2.set_xlim([0.0, 1.0])
ax2.set_xlabel('False Positive Rate')
ax2.set_ylabel('True Positive Rate')
ax2.set_title('Receiver Operating Characteristic (ROC) Curve')
ax2.legend(loc='center left', bbox_to_anchor=(1.05, 0.5), fontsize=8)
```
